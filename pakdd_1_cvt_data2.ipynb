```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os

import utl_eda
data_root = 'E:\\datasets\\tianchi\PAKDD\\'
```

# 读取数据

```python
df_logs = pd.DataFrame()
files = os.listdir(data_root + '\\hdf')
for hdf_file in files:
    if 'log_201' in hdf_file:
        filename_part = hdf_file.split('.')[0]
        r_file_path = data_root + 'hdf\\' + hdf_file
        print ('Reading {} .......'.format(r_file_path))
        df = pd.read_hdf(r_file_path)
        df_logs = pd.concat([df_logs, df])
```
