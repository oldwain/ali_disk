```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os

import utl_eda
data_root = 'E:\\datasets\\tianchi\PAKDD\\'
```

# 获取所有notnull列

```python
log_path = ['disk_sample_smart_log_2017', 'disk_sample_smart_log_2018_Q1','disk_sample_smart_log_2018_Q2']

dict_notnull = dict()
for d in log_path:
    files = os.listdir(data_root + d)
    for csv_file in files:
        
        filename_part = csv_file.split('.')[0]
        r_file_path = data_root + d + '\\' + csv_file
        print ('Reading {} .......'.format(r_file_path))
        chunks = pd.read_csv(r_file_path, index_col=0, nrows=None, chunksize=500000)
        for i, df in enumerate(chunks):

            ser_isnull = df.isnull().sum(axis=0)
            col_not_null = list(ser_isnull[ser_isnull != df.shape[0]].index)

            dict_notnull[csv_file + str(i)] = col_not_null
```

```python
len(dict_notnull.values() )
```

```python
col_set = set()

for cl in dict_notnull.values():
    col_set = col_set.union(set(cl))

keep_cols = sorted(list(col_set))
```
